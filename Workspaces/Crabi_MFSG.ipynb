        Funciones Nulos y tipo de Datos

```python
def report_nulls(df, name):
    """Imprime conteo y porcentaje de nulos por columna"""
    nulls = df.isna().sum()
    pct = (nulls / len(df)) * 100
    report = pd.DataFrame({
        "null_count": nulls,
        "null_pct": pct.round(2)
    })
    print(f"\n Valores nulos – {name}")
    print(report[report["null_count"] > 0])


def validate_dtypes(df, expected_types, name):
    """Valida tipos de datos esperados"""
    print(f"\n Validación de tipo de datos – {name}")
    for col, expected in expected_types.items():
        if col in df.columns:
            actual = df[col].dtype
            status = "OK" if actual == expected else "REVISAR"
            print(f"{col}: {actual} -- esperado {expected} [{status}]")
```

Se detecta birthdate como un campo crítico para el analisís etario, sin embargo con un porcentaje de datos mayor al 88% con fecha válida se puede realizar un analisís estadistico 

```python
import os
import re

FOLDER_PATH = r"C:\Users\sagua\OneDrive\Escritorio\Crabi"

# Palabras clave a buscar
KEYWORDS = [
    "daño material",
    "daños materiales",
    "robo total",
    "robo"
]

pattern = re.compile("|".join(KEYWORDS), re.IGNORECASE)

results = []

for file in os.listdir(FOLDER_PATH):
    if file.endswith(".xlsx"):
        file_path = os.path.join(FOLDER_PATH, file)
        try:
            xls = pd.ExcelFile(file_path)
        except Exception as e:
            print(f"No se pudo abrir {file}: {e}")
            continue

        for sheet in xls.sheet_names:
            try:
                df = pd.read_excel(xls, sheet_name=sheet)
            except Exception as e:
                print(f"Error leyendo {file} - {sheet}: {e}")
                continue

            # Solo columnas de texto
            for col in df.select_dtypes(include="object").columns:
                matches = (
                    df[col]
                    .dropna()
                    .astype(str)
                    .str.contains(pattern)
                )
                if matches.any():
                    results.append({
                        "file": file,
                        "sheet": sheet,
                        "column": col,
                        "matches": int(matches.sum())
                    })

results_df = pd.DataFrame(results)

results_df
```

Ninguna de estas columnas está normalizada ni vinculada directamente a coverage_id en service

Se identificaronn siniestros cuya descripcion sugiere robo total o daños materiales, sin embargo esta información no se encuentra normalizada ni vinculada a la cobertura del servicio

Dado que el cálculo de deducibles requiere una identificación precisa y consistente de la cobertura asociada a cada siniestro, esta información no fue considerada para el cálculo financiero.

## Limpieza de Datos

    claim

```python
# Eliminar duplicados
claim = claim.drop_duplicates(subset="id")

# Partner desde policy_number
claim["partner"] = claim["policy_number"].astype(str).str[:2]

# Eliminar registros sin ID
claim = claim[claim["id"].notna()]
```

    service


```python
# Eliminar duplicados
service = service.drop_duplicates(subset="id")

# Amount a numérico (ya se validó)
service["amount"] = pd.to_numeric(
    service["amount"], errors="coerce"
)

# Deductible a booleano
service["deductible"] = (
    service["deductible"]
    .fillna(False)
    .astype(bool)
)
```

    people

```python
# Eliminar duplicados
people = people.drop_duplicates(subset="id")

people["birthdate"] = pd.to_datetime(
    people["birthdate"], errors="coerce"
)

current_year = pd.Timestamp.today().year
people["age"] = current_year - people["birthdate"].dt.year

# Se filtran edades fuera de rangos razonables (>100)
people = people[(people["age"] >= 0) & (people["age"] <= 100)]
```

Integración de datos y construcción del dataset analítico

    claim + people

```python
claim_people = claim.merge(
    people,
    left_on="id",
    right_on="claim_id",
    how="inner",
    suffixes=("_claim", "_people")
)
print("Claims únicos:", claim["id"].nunique())
print("Claims en claim_people:", claim_people["id_claim"].nunique())
```

    claim_people + service

```python
full_data = claim_people.merge(
    service,
    left_on="id_people",
    right_on="people_id",
    how="left",
    suffixes=("", "_service")
)
print("Filas finales:", len(full_data))
print("Servicios con amount:", full_data["amount"].notna().sum())
```

El proceso de integración evidencia que no todos los siniestros cuentan con información completa a nivel persona o servicio con impacto financiero. En particular, solo una fracción de los servicios presenta un valor monetario (amount), condición indispensable para el cálculo de siniestralidad y severidad.

Esta reducción del universo analítico no representa una pérdida de información relevante, sino una depuración necesaria para garantizar consistencia metodológica y evitar la inclusión de registros sin impacto económico.

## Siniestralidad mensual 

    Servicios con impacto económico (Gastos)

```python
services_valid = full_data[
    (full_data["amount"].notna()) &
    (full_data["amount"] < 0)
].copy()

services_valid["gross_cost"] = services_valid["amount"].abs()

print(f"Servicios con gasto: {len(services_valid)}")
print(f"Gasto bruto total: ${services_valid['gross_cost'].sum():,.2f}")
```

    Asociación del gasto al mes del siniestro

```python
services_valid["created_at_claim"] = pd.to_datetime(
    services_valid["created_at_claim"],
    errors="coerce"
)

services_valid["month"] = (
    services_valid["created_at_claim"]
    .dt.tz_convert(None)
    .dt.to_period("M")
    .astype(str)
)
```

```python
services_valid.groupby("id")["gross_cost"].count().describe()
```

**Revisar si los datos vienen con un problema de escala

    Agregación mensual de gastos

```python
monthly_losses = (
    services_valid
    .groupby("month", as_index=False)
    .agg(
        total_losses=("gross_cost", "sum"),
        claims=("id_claim", "nunique")
    )
)

monthly_losses = monthly_losses.sort_values("month")
monthly_losses
```

    Cálculo de siniestralidad mensual

```python
MONTHLY_PREMIUM = 200_000

monthly_losses["premium"] = MONTHLY_PREMIUM

monthly_losses["loss_ratio"] = (
    monthly_losses["total_losses"] / monthly_losses["premium"]
)

monthly_losses
```

loss_ratio > 1 -- mes no rentable

loss_ratio < 1 -- mes rentable

El deducible solo se aplica cuando la cobertura está claramente identificada y normalizada. En los servicios donde no existe dicha identificación, el análisis se realiza sin deducible¿

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(monthly_losses["month"], monthly_losses["loss_ratio"])
plt.axhline(1, linestyle="--")
plt.xticks(rotation=45)
plt.title("Siniestralidad mensual (prima asumida)")
plt.xlabel("Mes")
plt.ylabel("Loss Ratio")
plt.tight_layout()
plt.show()
```

Bajo este supuesto de prima, el portafolio sería no rentable en todo el periodo

Octubre es un mes atípico

Pocos siniestros pueden distorsionar fuertemente el resultado mensual

## Cobertura con mayor y menor cantidad de siniestros

```python
claims_by_coverage = (
    full_data[
        full_data["coverage_id"].notna()
    ][["id_claim", "coverage_id"]]
    .drop_duplicates()
)
coverage_counts = (
    claims_by_coverage
    .groupby("coverage_id", as_index=False)
    .agg(
        claims=("id_claim", "nunique")
    )
)
max_coverage = coverage_counts.loc[
    coverage_counts["claims"].idxmax()
]

min_coverage = coverage_counts.loc[
    coverage_counts["claims"].idxmin()
]

print("Cobertura con más siniestros:")
print(max_coverage)

print("\nCobertura con menos siniestros:")
print(min_coverage)
```

```python
coverage_counts.sort_values("claims", ascending=False).plot(
    x="coverage_id",
    y="claims",
    kind="bar",
    title="Cantidad de siniestros por cobertura",
    legend=False
)
```

## Métricas por Partner

```python
partner_metrics.sort_values("total_cost", ascending=False).plot(
    x="partner",
    y="total_cost",
    kind="bar",
    title="Siniestralidad total por Partner",
    legend=False
)
```

Se detectaron 57 eventos asociados a conductores menores de edad, los cuales presentan una severidad promedio significativamente superior al resto del portafolio, por lo que se recomienda un análisis específico de estos casos.

        Analisís Etario

```python
bins = [0, 25, 35, 45, 55, 65, 100]
labels = ["<25", "25-34", "35-44", "45-54", "55-64", "65+"]

services_valid["age_group"] = pd.cut(
    services_valid["age"],
    bins=bins,
    labels=labels
)
```

```python
age_metrics = (
    services_valid
    .groupby("age_group", observed=True)
    .agg(
        claims=("id_claim", "nunique"),
        avg_severity=("gross_cost", "mean"),
        total_severity=("gross_cost", "sum")
    )
    .reset_index()
)

age_metrics
```

Del total de 569 siniestros, solo 173 tienen servicios con gasto registrado, lo que representa alrededor del 30% del total. Esto indica que una parte significativa de los reclamos no impacta financieramente al portafolio, ya sea por falta de servicios asociados o por registros incompletos. Este hallazgo es importante para contextualizar la siniestralidad mensual, las métricas por partner y cobertura, y las comparaciones por rango etario, ya que los análisis financieros solo consideran los claims con gasto real
